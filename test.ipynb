```python
numerical_features=[col for col in df.columns if df[col].dtype!='O']
print(f"there are {len(numerical_features)} numerical_features :  {numerical_features}")

cateorical_features=[col for col in df.columns if df[col].dtype=='O']
print(f"there are {len(cateorical_features)} categorical features : {cateorical_features}")
```

```python
plt.figure(figsize=(15,10))
plt.suptitle("Univarient analysis on numerical features", fontsize=20)

for i in range(len(numerical_features)):
    plt.subplot(3,3,i+1)
    sns.kdeplot(x=df[numerical_features[i]],color='r',shade=True)
    plt.tight_layout()
```

```python
plt.figure(figsize=(15,10))
plt.suptitle("Univarient analysis of numerical feature")

for i in range(len(numerical_features)):
    plt.subplot(3,3,i+1)
    sns.kdeplot(x=df[numerical_features[i]],color='b',shade=True)
    plt.tight_layout()
```

```python
plt.figure(figsize=(15,10))
plt.suptitle("Outlier detection")

for i in range(len(numerical_features)):
    plt.subplot(3,3,i+1)
    sns.boxplot(x=df[numerical_features[i]],color='g')
    plt.tight_layout()
```

```python
plt.figure(figsize=(15,10))
plt.suptitle("univarient analysis on categorical features")

for i in range(len(cateorical_features)):
    plt.Subplot(6,1,i+1)
    sns.countplot(x=df[cateorical_features[i]],order=df[cateorical_features[i]].value_counts().index,palette='set1',ec='black',linewidth=1.5)
    plt.tight_layout()
    plt.xticks(rotation=90)
```
